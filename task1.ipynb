```python
# Task 1.3: Attempt revenue calculation (Revenue = Quantity * Price Per Unit)

df['Revenue'] = df['Quantity'] * df['Price Per Unit']

# Why did it fail? The 'Price Per Unit' column contains string values with a dollar sign,
# which cannot be directly multiplied with the 'Quantity' column.
# We need to clean the 'Price Per Unit' column by removing the dollar sign
# and converting it to a numeric type before performing the calculation.
```

```python
# Task 1.4: Compute total revenue per item (Hint use: groupby)
# Show the output. Describe what looks wrong. (Hint: Inspect the data type of Total Spent.)

total_revenue_per_item = df.groupby('Item')['Total Spent'].sum()
print(total_revenue_per_item)
print()
print(df['Total Spent'].dtype)

# The Total Spent olumn contains ERROR string values mixed in with numeric values.
# Because of this, pandas infers the dtype as object (string), insteaf of float or int.
# This causes .sum() to concatentate the strings instead of adding them.
```

```python
# Task 1.5: Try to extract Transaction month and Day of week using Transaction Date. (Hint: Try using
# .dt.month and .dt.day_name(). You might need datetime module.)
# If you get an error, show the error message. Explain why it occurred.

print("Data type of 'Transaction Date':", df['Transaction Date'].dtype)
print()

try:
    df['Transaction Month'] = df['Transaction Date'].dt.month
    df['Transaction Day'] = df['Transaction Date'].dt.day_name()
    print(df[['Transaction Date', 'Transaction Month', 'Transaction Day']].head())
except AttributeError as e:
    print("AttributeError:", e)

# Explanation:
# The 'Transaction Date' column contains "ERROR" string values mixed in with valid date strings.
# Because of these non-date entries, pandas reads the column as dtype 'object' (string)
# rather than 'datetime64'. The .dt accessor is only available on datetime columns, so
# attempting to use .dt.month or .dt.day_name() on an object column raises an AttributeError.
# The column must first be cleaned (removing/replacing "ERROR" entries) and then converted
# with pd.to_datetime() before datetime properties can be extracted.
```
