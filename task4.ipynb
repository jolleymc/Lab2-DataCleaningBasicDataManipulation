```python
import pandas as pd

df = pd.read_csv('cafe_sales.csv')
```

```python
# Task 4.1
missing_percentage = df.isnull().mean() * 100
print(missing_percentage)
```

```python
# Task 4.2
import pandas as pd
import numpy as np

df = df.replace(['UNKNOWN', 'ERROR'], np.nan)

cols_to_numeric = ['Quantity', 'Price Per Unit', 'Total Spent']
for col in cols_to_numeric:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
for col in cols_to_numeric:
    df[col] = df[col].fillna(df[col].median())
    
print(df)
```

```python
# Task 4.3
# Nature of missingness: Quantity and Price Per Unit appear to be MCAR; Item and Payment Method are likely MAR; Location is likely MNAR

# Justification methods: Median Imputation (Numerical): I chose the Median for Quantity and Price because it is robust to outliers. The Median
# is not pulled away by extreme values. Mode Imputation (Categorical): For Location and Item, I used the Mode because these are non-numeric
# categories. We cannot calculate an average location, so filling with the most frequent value is the standard approach to maintain the
# largest possible dataset for analysis.

# Bias: Variance Reduction: By filling thousands of cells with the exact same median or mode value, we reducing the variety in the data. Selection
# Bias: Using the Mode for Location creates a risk of Selection Bias. If the 3,265 missing rows actually belonged to a hidden location, 
# attributing them all to the most popular location (e.g., "Takeaway") will lead to an incorrect conclusion

# Risks: If chosen to Drop Rows instead of imputing, we would have lost over 30% of the dataset. This would lead to a significant underestimation
# of total revenue and likely produce cleaner looking but highly misleading results.

# Impact of downstream analysis: These decisions ensure that the code in Section 5 will run without errors (no NaN math errors).
# However, the analyst must remember that the "Top Location" or "Average Quantity" results are now influenced by the imputation.
```
