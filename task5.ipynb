```python
import pandas as pd

df = pd.read_csv('cafe_sales.csv')
```

```python
# Task 5.1
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')

df['Revenue'] = df['Quantity'] * df['Price Per Unit']

df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')

df['Transaction_Month'] = df['Transaction Date'].dt.month
df['Day_of_Week'] = df['Transaction Date'].dt.day_name()

print(df[['Quantity', 'Price Per Unit', 'Revenue']].dtypes)
```

```python
# Task 5.2 
revenue_per_item = df.groupby('Item')['Revenue'].sum()
print("Total Revenue per Item:")
print(revenue_per_item)
print("-" * 30)

top_3_items = revenue_per_item.sort_values(ascending=False).head(3)
print("Top 3 Revenue-Generating Items:")
print(top_3_items)
print("-" * 30)

revenue_per_location = df.groupby('Location')['Revenue'].sum()
print("Total Revenue per Location:")
print(revenue_per_location)
```
